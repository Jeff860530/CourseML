```
import matplotlib.pyplot as plt
import numpy as np
plt.ion()
#create the first matrix ,also need to modify the shape of result_matrix
#both matrix are in same size
matrix = np.random.randint(0,2, size=(20,20))
print("matrix:\n",matrix)
test_matrix = np.array([[1, 1, 1],[1, 0, 1],[1, 1, 1]])
print("test_matrix:\n",test_matrix)
#iniliat the result_matrix
result_matrix = np.random.randint(0,1, size=(20,20))
print("result_matrix:\n",result_matrix)
```

```
def game_of_live(matrix):
    padding_matrix = np.pad(matrix, 1, 'constant', constant_values = 0)
    #print("padding_matrix:\n",padding_matrix)
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            cross_matrix = padding_matrix[i:i+3, j:j+3] * test_matrix
            #print("result_matrix:\n",cross_matrix)
            #print("result_matrix.sum():",cross_matrix.sum())
            if (cross_matrix.sum() < 2):
                result_matrix[i,j] = 0
            elif (cross_matrix.sum() == 2 or cross_matrix.sum() == 3):
                result_matrix[i,j] = 1
            elif (cross_matrix.sum() > 3):
                result_matrix[i,j] = 0
    return result_matrix
```

```
for i in range (100):
    result_matrix = game_of_live(matrix)
    #print(result_matrix)
    matrix = result_matrix
    plt.imshow(matrix)
    plt.show()
```
